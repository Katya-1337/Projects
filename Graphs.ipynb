Импорт библиотек

```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go

!pip install yfinance
import yfinance as yf
```

Загрузка 6 датасетов с YFinance

```python
# Список тикеров
tickers_list = ['AAPL', 'WMT', 'IBM', 'MU', 'BA', 'AXP'] #Apple, Walmart, International Business Machines, Micron Technology, The Boeing, American Express

#массив с дата-сетами
df_set = []

# Цикл для загрузки данных о цене акций для каждой акции
for symbol in tickers_list:
    ticker = yf.Ticker(symbol)
    df = ticker.history(start = "2022-05-15", end = "2023-05-15", interval = "1wk").reset_index()
    df_set.append(df)
```

Графики Line и Candle

```python
#цикл создания графиков line и candle для каждого датасета
for symbol in tickers_list:
    #line
    fig = px.line(df,x = 'Date',y = ['Open','Close'],title='Изменение цены открытия и цены закрытия '+str(symbol))
    fig.show()

    #candle
    fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                                      open=df['Open'],
                                      high=df['High'],
                                      low=df['Low'],
                                      close=df['Close'])])
    fig.show()
```

```python
#Соединения данных компаний взятых с yfinance в таблицу
data = {'date':list(df_set[0]['Date']),
        'AAPL' : list(df_set[0]['Open']),
        'WMT': list(df_set[1]['Open']),
        'IBM':list(df_set[2]['Open']),
        'MU':list(df_set[3]['Open']),
        'BA':list(df_set[4]['Open']),
        'AXP':list(df_set[5]['Open'])}

new_df = pd.DataFrame(data)
new_df
```

Построение графика Area

```python
fig = px.area(new_df,x= 'date',y = ['AAPL', 'WMT', 'IBM', 'MU', 'BA', 'AXP'],title = 'Изменение цены')
fig.show()
```

Загрузка 2 датасетов c investing.com

```python
import pandas as pd

#Загрузка файлов
binance = pd.read_csv('/content/Binance USD Historical Data - Investing.com.csv')
ethereum = pd.read_csv("/content/Ethereum Historical Data - Investing.com.csv")
```

Построение графика Bar

```python
#редактирование значений
binance['Change %'] = binance['Change %'].apply(lambda x: float(str(x).replace('%','').replace(',','.')))
ethereum['Change %'] = ethereum['Change %'].apply(lambda x: float(str(x).replace('%','').replace(',','.')))

#График для binance
fig = px.bar(binance, x = 'Date', y = 'Change %', title = 'binance')
fig.show()

#График для ethereum
fig = px.bar(ethereum, x = 'Date', y = 'Change %', title = 'ethereum')
fig.show()
```
